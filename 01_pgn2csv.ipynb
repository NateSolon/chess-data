```python
#export
import chess.pgn
import pandas as pd
```

```python
#export
def pgn2df(pgn):
    game = chess.pgn.read_game(pgn)
    game_info = []
    while game:
        game_info.append(dict(game.headers))
        game = chess.pgn.read_game(pgn)
    df = pd.DataFrame(game_info)
    return df
```

```python
pgn = open("data/magnus.pgn")
```

```python
df = pgn2df(pgn)
```
